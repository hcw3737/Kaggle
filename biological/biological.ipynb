```python
train = pd.read_csv("/kaggle/input/bioresponse/train.csv")
test = pd.read_csv("/kaggle/input/bioresponse/test.csv")
```

```python
display(train, test)
```

```python
all_data = pd.concat([train, test])
all_data2 = all_data.drop(["Activity"],1)
```

```python
train2 = all_data2[:len(train)]
test2 = all_data2[len(train):]
```

```python
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
for i in [3,4,5,6,7,8,9]:
    xgb = XGBClassifier(learning_rate=0.1, max_depth=i)
    print(i)
    print(-cross_val_score(xgb, train2, train["Activity"], n_jobs=4, cv=5, scoring='neg_log_loss').mean())
```

```python
for i in [0.5, 0.7, 0.8,0.9, 1]:
    xgb = XGBClassifier(learning_rate=0.1, max_depth=6,colsample_bytree=i)
    print(i)
    print(-cross_val_score(xgb, train2, train["Activity"], n_jobs=4, cv=5, scoring='neg_log_loss').mean())
```

```python
for i in [0.5, 0.7, 0.8,0.9, 1]:
    xgb = XGBClassifier(learning_rate=0.1, max_depth=6,colsample_bytree=0.5 ,subsample=i)
    print(i)
    print(-cross_val_score(xgb, train2, train["Activity"], n_jobs=4, cv=5, scoring='neg_log_loss').mean())
```

```python
for i in [70,0,90,100,120]:
    xgb = XGBClassifier(learning_rate=0.1, max_depth=6,colsample_bytree=0.5 ,subsample=1, n_estimator=i)
    print(i)
    print(-cross_val_score(xgb, train2, train["Activity"], n_jobs=4, cv=5, scoring='neg_log_loss').mean())
```

```python
xgb = XGBClassifier(learning_rate=0.01, max_depth=6,colsample_bytree=0.5 ,subsample=1, n_estimator=1000)
print(0.01)
print(-cross_val_score(xgb, train2, train["Activity"], n_jobs=4, cv=5, scoring='neg_log_loss').mean())
```

```python
xgb = XGBClassifier(learning_rate=0.02, max_depth=6,colsample_bytree=0.5 ,subsample=1, n_estimator=500)
print(0.02)
print(-cross_val_score(xgb, train2, train["Activity"], n_jobs=4, cv=5, scoring='neg_log_loss').mean())
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(learning_rate=0.1, max_depth=6,colsample_bytree=0.5 ,subsample=1, n_estimator=100)
# %%time
xgb.fit(train2, train["Activity"])
result = xgb.predict_proba(test2)
```

```python
sub = pd.read_csv("/kaggle/input/bioresponse/svm_benchmark.csv")
sub["PredictedProbability"] = result[:,1]
sub.to_csv("sub0913.csv", index=0)
```
