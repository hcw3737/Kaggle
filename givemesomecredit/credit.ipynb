# 모두연 실습 0123

```python
#train과 test 데이터를 합쳐서 한번에 데이터 타입 변경 편리하도록 합친다.
all_data = pd.concat([train,test])
```

```python
display(train,test)
```

```python
train.isnull().sum()
#null값이 몇개씩 있는지 확인(각 컬럼마다)
```

```python
#randomforest를 돌릴때는 결측치를 제거해줘야한다.
all_data = all_data.fillna(-1)

#0말고 -1 같은 걸 넣어주면, 약간 예외값처럼 학습을 한다.
```

```python
all_data.isnull().sum()
```

```python
# all_data2 = all_data.drop(('SeriousDlqin2yrs'),1)   #아래와 동일
all_data2 = all_data.drop(['Unnamed: 0','SeriousDlqin2yrs'], axis=1)
train2
# SeriousDlqin2yrs 가 타겟변수
# Unnamed 인덱스 칼럼인듯, 중복되서 필요X?
```

```python
# len(train)
train2 = all_data2[:len(train)]
test2 = all_data2[len(train):]
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(train2,train["SeriousDlqin2yrs"])
result = rf.predict_proba(test2)

#그냥 predict는 0,1로만 나온다
#predict_proba : 확률로 나옴
#분류문제는 확률값으로 제출해야한다.
```

```python
#결과 확인
result
```

```python
rf.classes_
# result의 왼쪽 확률은 0이 나올 확률
# result의 오른쪽 확률은 1이 나올 확률
```

```python
sub = pd.read_csv('/kaggle/input/GiveMeSomeCredit/sampleEntry.csv')
```

```python
sub['Probability'] = result[:,1]  
# 2번째 열 확률부분만 가져와야한다.
sub.to_csv('submission.csv',index=0)
```
