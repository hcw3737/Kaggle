```python
import pandas as pd

path = '/content/drive/MyDrive/Colab Notebooks/Dacon/open/'
test_df = pd.read_csv(path+"test_data.csv")
```

```python
test_df
```

```python
premise, hypothesis = list(test_df['premise']), list(test_df['hypothesis'])
```

### pororo 라이브러리를 활용하여 제출할 결과 만들기

```python
from pororo import Pororo
```

```python
nli_module = Pororo(task="nli", lang="ko")
```

```python
nli_module(premise[0], hypothesis[0])
```

### 진행상황을 확인할때 사용할 tqdm 라이브러리 설치

```python
!pip install tqdm
```

```python
from tqdm import tqdm_notebook

nli_results = []

for pre, hypo in tqdm_notebook(zip(premise, hypothesis)):
  result = nli_module(pre, hypo)

  nli_results.append(result)
```

### 제출할 결과 파일 만들기

```python
sub_df = pd.read_csv(path+"sample_submission.csv")
sub_df['label'] = nli_results
```

```python
sub_df.to_csv("./pororo_submission.csv", index=False)
```
